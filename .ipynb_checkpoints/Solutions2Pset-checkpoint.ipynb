```python
import Agent
import Env
import Policy
import State
import time
import numpy as np
```

```python
# initialize the state, policy, discount and agent
agent = Agent.Agent(6,6,0)
policy = Policy.Policy()
init_state = State.State(1, 4, 6)
discount = 0.9
```

```python
# Prob 3(c)
agent.plot_trajectory(policy, init_state, 0)
```

```python
# Prob 3(e)
value = agent.policy_eval(policy, 0.9)
print("The value of trajectory in 3(c) is, ", value[6][1][4])
```

```python
# Prob 3(h) 3(i)
start_time = time.time()

opt_policy3, opt_value3 = agent.policy_iter(policy, discount)

end_time = time.time()

diff_time = end_time - start_time

print("3(i) running time: ", diff_time, "s")

agent.plot_trajectory(opt_policy3, init_state, 0)

print("The optimal value of 3(h) is: ", opt_value3)
```

```python
init_x, init_y, init_dir = init_state.getState()
opt_value3[init_dir][init_x][init_y]
```

```python
# Prob 4(b) 4(c)
start_time = time.time()

opt_policy4, opt_value4 = agent.value_iter(discount)

end_time = time.time()

diff_time = end_time - start_time

print ("4(c) running time: ", diff_time, "s")

agent.plot_trajectory(opt_policy4, init_state, 0)
```

```python
opt_value4[init_dir][init_x][init_y]
```

```python
# Prob 5(a)
pe = 0.25
traj_5a = agent.plot_trajectory(policy, init_state,pe)
traj_value_5a = agent.traj_value(traj_5a)
print("The value of trajectory in 5(a) is, ", traj_value_5a)
```

```python
# 5(b) Modified reward function
traj_5b = agent.plot_trajectory_mod(policy, init_state, pe)
traj_value_5b = agent.traj_value_mod(traj_5b)
print("The value of trajectory in 5(a) is, ", traj_value_5b)
```
